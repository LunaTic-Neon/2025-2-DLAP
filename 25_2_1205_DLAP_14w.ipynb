```python
# BERT
# 라이브러리 설치
!pip uninstall -y torchvision torchaudio

!pip install torch==2.2.0 --index-url https://download.pytorch.org/whl/cpu
!pip install "transformers>=4.41, <5.0"

# 라이브러리 호출
import pandas as pd
import numpy as np
import tensorflow as tf
import re, random
from transformers import BertTokenizer

# 1. IMDB 데이터 로드
print("IMDB 데이터 로드 시작")
movie_reviews = pd.read_csv(
    "/content/drive/MyDrive/Data file/DLAP 14 실습에 필요한 자료들-20251205/IMDB Dataset.csv")

# 2. 텍스트 전처리 함수
def preprocess_text(sen):
  sen = re.sub("<.*?>", ' ', sen)
  sen = re.sub("[^a-zA-Z]", " ", sen)
  sen = re.sub(r"\s+[a-zA-Z]\s+", ' ', sen)
  sen = re.sub(r'\s+', ' ', sen)
  return sen.strip()

print("텍스트 전처리 시작")
reviews = [preprocess_text(t) for t in movie_reviews['review']]

# 3. 레이블 변환

print("레이블 변환 시작")
y = np.array([1 if s == "positive" else 0 for s in movie_reviews["sentiment"]])

# 4. HugginFace BERT 토크나이저
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def tokenize_sentence(sent):
  return tokenizer.encode(
      sent,
      add_special_tokens=True,
      max_length=512,
      truncation=True
  )

print("토큰화 수행")
tokenized_reviews = [tokenize_sentence(r) for r in reviews]

pairs = list(zip(tokenized_reviews, y))

# 5. TF Dataset 생성

# 전체 길이
TOTAL = len(pairs)
TEST = TOTAL // 10    # 10%
TRAIN = TOTAL - TEST

train_pairs =pairs[:TRAIN]
test_pairs = pairs[TRAIN:]

# generator
def gen_train():
  for seq, label in train_pairs:
    yield tf.constant(seq, dtype=tf.int32), tf.constant(label, dtype=tf.int32)

def gen_test():
  for seq, label in test_pairs:
    yield tf.constant(seq, dtype=tf.int32), tf.constant(label, dtype=tf.int32)

train_dataset = tf.data.Dataset.from_generator(
    gen_train,
    output_signature=(
        tf.TensorSpec(shape=(None,), dtype=tf.int32),
        tf.TensorSpec(shape=(), dtype=tf.int32)
  )
)

test_dataset = tf.data.Dataset.from_generator(
    gen_test,
    output_signature=(
        tf.TensorSpec(shape=(None,), dtype=tf.int32),
        tf.TensorSpec(shape=(), dtype=tf.int32)
  )
)

BATCH_SIZE = 32

train_data = train_dataset.shuffle(5000).padded_batch(
    BATCH_SIZE,
    padded_shapes=([None], ())
).prefetch(tf.data.AUTOTUNE)

test_data = test_dataset.padded_batch(
    BATCH_SIZE,
    padded_shapes=([None], ())
).prefetch(tf.data.AUTOTUNE)

# 6. CNN 감성 분석 모델 정의
class TEXT_MODEL(tf.keras.Model):
  def __init__(self, vocab_size, embed_dim=128, can_filters=100, dnn_units=256, dropout_rate=0.2):
    super().__init__()

    self.embedding = tf.keras.layers.Embedding(vocab_size, embed_dim)

    self.conv2 = tf.keras.layers.Conv1D(filters=cnn_filters, kernal_size=2, activation="relu")
    self.conv3 = tf.keras.layers.Conv1D(filters=cnn_filters, kernal_size=3, activation="relu")
    self.conv4 = tf.keras.layers.Conv1D(filters=cnn_filters, kernal_size=4, activation="relu")

    self.pool = tf.keras.layers.GlobalMaxPool1D()

    self.dense = tf.keras.layers.Dense(dnn_units, activation="relu")
    self.dropout = tf.keras.layers.Dropout(dropout_rate)

    self.output_layer = tf.keras.layers.Dense(1, activation="sigmoid")

  def call(self, x, training=False):
    x = self.embedding(x)

    x2 = self.pool(self.conv2(x))
    x3 = self.pool(self.conv3(x))
    x4 = self.pool(self.conv4(x))

    x = tf.concat([x2, x3, x4], axis=-1)
    x = self.dense(x)
    x = self.output(x, training=training)
    return self.output_layer(x)

# 7. 모델 학습
VOCAB_SIZE = tokenizer.vocab_size

model = TEXT_MODEL(VOCAB_SIZE)

model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)

print("모델 학습 시작")
model.fit(train_data, epochs=3)

# 8. 모델 평가
print("모델 평가:")
results = model.evaluate(test_data)
print("Test Loss, Test Accuracy =", results)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

text = """과일 가게에 사과가 많이 진열괴어 있다
그녀가 나에게 사과한 후, 우리는 친해졌다
애플은 사과 모양을 로고로 사용한다\n"""
```

```python
tok = Tokenizer()
tok.fit_on_texts([text])

vocSize = len(tok.word_index) + 1

seqs = list()
for word in text.split("\n"):
  encoded = tok.texts_to_sequences([word])[0]
  for i in range(1, len(encoded)):
    seq = encoded[:i+1]
    seqs.append(seq)

maxLen = max(len(i) for i in seqs)

seqs = pad_sequences(seqs, maxlen=maxLen, padding="pre")
seqs
```

```python
seqs = np.array(seqs)
x = seqs[:,:-1]
```

```python
from tensorflow.keras.utils import to_categorical
y = seqs[:, -1]
y = to_categorical(y, num_classes = vocSize)
```

```python
from tensorflow.keras.layers import LSTM
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.layers import Embedding

model = Sequential()
model.add(Embedding(vocSize, 10, input_length=maxLen-1))
model.add(LSTM(32))
model.add(Dense(vocSize, activation="softmax"))
model.compile(loss="categorical_crossentropy", metrics=["accuracy"],
              optimizer="adam")
model.fit(x, y, epochs=200)
```

```python
def sentGen(model, tok, word, n):
  sent = ""
  word2 = word

  for _ in range(n):
    encoded = tok.texts_to_sequences([word])[0]
    encoded = pad_sequences([encoded], maxlen=7, padding="pre")

    pred = model.predict(encoded)
    res = np.argmax(pred, axis=1)[0]

    for w, i in tok.word_index.items():
      if i == res:
        next_word = w
        break

    word = word + " " + next_word
    sent = sent + " " + next_word

  return word2 + sent
```
